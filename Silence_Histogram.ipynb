# Calculate lengths of the silent periods of an audio file

## Poetry Reading

```python
from librosa.core import audio
import numpy as np
import pickle
import matplotlib.pyplot as plt
import pandas as pd

# Load DataFrame
pkl_path = '/content/gdrive/My Drive/Papers and Projects/MultimodalPoem/CSV files/poetry_more_metadata_small_first_row.pkl'
row = pd.read_pickle(pkl_path)

transcriptions = row['transcriptions_list']

# Initialize empty array to store silence durations
silence_durations = []

# Initialize variable for the last end time
last_end_time = 0.0

# Loop through segments to find silence durations, excluding the first and last silences
for i, segment in enumerate(transcriptions["segments_after_alignment"]):
    for j, word_info in enumerate(segment["words"]):
        if "start" in word_info and "end" in word_info:
            start_time = word_info["start"]
            end_time = word_info["end"]

            # For the first word, we don't want to consider the silence before it
            if i == 0 and j == 0:
                last_end_time = end_time
                continue

            # Calculate silence duration
            silence_duration = start_time - last_end_time

            # Add the silence duration to the array
            silence_durations.append(silence_duration)

            # Update last_end_time
            last_end_time = end_time

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(silence_durations, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Silence Durations of Poetry Reading Audio')
plt.xlabel('Duration')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.6)
plt.tight_layout()
plt.show()
```
